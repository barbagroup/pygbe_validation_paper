```python
import numpy
from matplotlib import pyplot, rcParams
%matplotlib notebook
```

```python
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
rcParams['axes.linewidth'] = 1
```

```python
AR_black, w_black = numpy.loadtxt('ellis_figS4_digit/black_curve.csv', delimiter=',', unpack=True)
AR_red, w_red = numpy.loadtxt('ellis_figS4_digit/red_curve.csv', delimiter=',', unpack=True)
AR_sim, w_sim = numpy.loadtxt('AR_lower_modes.txt', skiprows=1, usecols=[0,1],unpack=True)
```

```python
pyplot.figure(figsize=(8, 4))

pyplot.plot(AR_black, w_black, '-o', c='black', label='Ellis 2016 (5000 nm gap)')
#pyplot.plot(AR_red, w_red, '-o', c='red',alpha=0.3, label='Ellis 2016 - 500 nm gap - coupled pillars ')
pyplot.plot(AR_sim, w_sim, '-x', c='green', label='PyGBe')


pyplot.tick_params(direction='in')
pyplot.xticks(numpy.arange(0., 7.5, step=1), rotation=25)
pyplot.yticks(numpy.arange(800, 1000, step=20), rotation=25)
pyplot.xlabel('Aspect ratio L/W ')
pyplot.ylabel('Wave number [$cm^{-1}$]')
pyplot.xlim(0.5, 7.5)
pyplot.ylim(800, 990)
pyplot.grid(True)
pyplot.legend();
pyplot.savefig('AR_rep_val_figs/AR_rep_FS4_Ellis2016.pdf', format='pdf', bbox_inches='tight', pad_inches=0.1)
```
