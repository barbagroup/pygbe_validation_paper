Comparison of short and long edge configuration when we extend the third-direction (y) from 13440 to 26880 Angstroms.

**Short edge configuration** means that the light is on the direction of `x=6720` Angstroms and therefore the Electric field pointing parallel to the short edge `z=3280` Angstrom

**Long edge configuration** means that the light is on the direction of `x=3280` Angstroms and therefore the Electric field pointing parallel to the short edge `z=6720` Angstrom. 

```python
import numpy
from matplotlib import pyplot, rcParams
%matplotlib inline
```

```python
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 20
rcParams['axes.linewidth'] = 1
```

Note: this result was not included on the paper "Reproducible Validation and Replication Studies in Nanoscale Physics"since it is an intermediate result. 

```python
#Data from y=13440 Ang

w_SE, cext_SE = numpy.loadtxt('results_data/prism_672x1344x328nm_res/prism_36K_short_edge10-20microns.txt',
                        skiprows=1, unpack=True)
w_LE, cext_LE = numpy.loadtxt('results_data/prism_672x1344x328nm_res/prism_36K_long_edge10-20microns.txt',
                        skiprows=1, unpack=True)
```

```python
#Data from y=13440 Ang

w_SE_reg_1, cext_SE_reg_1 = numpy.loadtxt('results_data/prism_regular_SE_LE_res/prism_reg_36K_SE_y13K10-20microns.txt',
                        skiprows=1, unpack=True)
w_LE_reg_1, cext_LE_reg_1 = numpy.loadtxt('results_data/prism_regular_SE_LE_res/prism_reg_36K_LE_y13K10-20microns.txt',
                        skiprows=1, unpack=True)
```

```python
#plot
pyplot.figure(figsize=(8, 6))
pyplot.plot((w_SE/10000), cext_SE/cext_SE.max(), label='$y=13440\;\AA$', lw=2)
pyplot.plot((w_SE_reg_1/10000), cext_SE_reg_1/cext_SE_reg_1.max(), label='$y=13440 \;\AA$ reg ', lw=2)

pyplot.xlabel('$\lambda [\mu m]$')
pyplot.ylabel('$C_{ext} $ / max_in_each_curve')
pyplot.title('Short Edge compare to Fig. 14a in paper')
pyplot.grid(True)
pyplot.xticks(rotation=45)

pyplot.legend();
```

```python
#plot
pyplot.figure(figsize=(8, 6))
pyplot.plot((w_LE/10000), cext_LE/cext_LE.max(), label='$y=13440 \;\AA$', lw=2)
pyplot.plot((w_LE_reg_1/10000), cext_LE_reg_1/cext_LE_reg_1.max(), label='$y=13440 \;\AA$ reg', lw=2)

pyplot.xlabel('$\lambda [\mu m]$')
pyplot.ylabel('$C_{ext} $ / max_in_each_curve')
pyplot.title('Long Edge compare to Fig. 14b in paper')
pyplot.grid(True)
pyplot.xticks(rotation=45)

pyplot.legend();
```
