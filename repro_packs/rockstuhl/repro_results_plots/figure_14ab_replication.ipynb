## Replication of Figure 14 of Rockstuhl et al. 2005 

This notebook contains the results of the replication of Figure 14 of Rockstuhl et al. 2005, compared with its digitized data.

```python
import numpy
from matplotlib import pyplot, rcParams
%matplotlib inline
```

```python
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 20
rcParams['axes.linewidth'] = 1
```

```python
#Data from y=26680 Ang regular and round mesh

w_SE_round, cext_SE_round = numpy.loadtxt('results_data/prism_round_reg_SE_LE_res/prism_round_reg_38K_short_edge10-20microns.txt',
                                  skiprows=1, unpack=True)
w_LE_round, cext_LE_round = numpy.loadtxt('results_data/prism_round_reg_SE_LE_res/prism_round_reg_38K_long_edge10-20microns.txt',
                        skiprows=1, unpack=True)
```

```python
w_14a, c_14a = numpy.loadtxt('figure_14_rocksthul_digitized/short_edge.csv', delimiter=',', unpack=True)
w_14b, c_14b = numpy.loadtxt('figure_14_rocksthul_digitized/long_edge.csv', delimiter=',', unpack=True)
```

 ## Figure 14a

```python
from scipy.signal import argrelextrema
```

```python
idx_peaks_SE = argrelextrema(cext_SE_round, numpy.greater)[0]
w_SE_peaks_pygbe = w_SE_round[idx_peaks_SE]/10000

w_SE_peaks_pygbe
```

### Rockstuhl 2005 short edge

They report their peaks at 10.42 and 10.7 microns for the short edge configuration


```python
#plot
pyplot.figure(figsize=(8, 6))

pyplot.plot((w_SE_round/10000), cext_SE_round/cext_SE_round.max(), label='PyGBe', lw=2, c='C2')

for pp in [10.403, 10.716]:
    pyplot.axvline(x=pp, c='C2',alpha=0.7, ls='--')
    pyplot.text(x=pp, y=0.4, s=f'{pp:.2f}' , rotation=45, c='C2', alpha=0.7)


pyplot.plot(w_14a, c_14a, label='Rockstuhl', lw=2, c='C1')

for pw in [10.42, 10.7]:
    pyplot.axvline(x=pw, c='C1',alpha=0.7, ls='--')
    pyplot.text(x=pw, y=0.8, s=f'{pw:.2f}' , rotation=45, c='C1', alpha=0.7)

pyplot.xlabel('$\lambda [\mu m]$')
pyplot.xticks(rotation=45)
pyplot.ylabel('$C_{ext} $/ max')
#pyplot.title('Short Edge compare to Fig. 14a in paper')
pyplot.title('$ E \parallel b$')

pyplot.grid(True)
pyplot.legend();
pyplot.savefig('replication_fig14/replication_14a.pdf', format='pdf',bbox_inches='tight');
```

## Figure 14b

```python
idx_peaks_LE = argrelextrema(cext_LE_round, numpy.greater)[0]
w_LE_peaks_pygbe = w_SE_round[idx_peaks_LE]/10000

w_LE_peaks_pygbe
```

### Rockstuhl 2005 Long edge

They report their peaks at 10.42 and 10.82 microns for the long edge configuration


```python
#plot
pyplot.figure(figsize=(8, 6))

pyplot.plot((w_LE_round/10000), cext_LE_round/cext_LE_round.max(), label='PyGBe', lw=2, c='C2')

for pp in [10.424, 10.828]:
    pyplot.axvline(x=pp, c='C2',alpha=0.7, ls='--')
    pyplot.text(x=pp, y=0.4, s=f'{pp:.2f}' , rotation=45, c='C2', alpha=0.7)


pyplot.plot(w_14b, c_14b, label='Rockstuhl',lw=2, c='C1')

for pw in [10.42, 10.82]:
    pyplot.axvline(x=pw, c='C1',alpha=0.7, ls='--')
    pyplot.text(x=pw, y=0.8, s=f'{pw:.2f}' , rotation=45, c='C1', alpha=0.7)


pyplot.xlabel('$\lambda [\mu m]$')
pyplot.xticks(rotation=45)
pyplot.ylabel('$C_{ext} $ / max')
#pyplot.title('Long Edge compare to Fig. 14b in paper')
pyplot.title('$ E \parallel a$')

pyplot.grid(True)
pyplot.legend();
pyplot.savefig('replication_fig14/replication_14b.pdf', format='pdf',bbox_inches='tight');
```
