## Grid independence for cube (L=535 nm)

The main result is in the last figure, which will save as a pdf under the directory `grid_indep_fig`. You can find the original plot that appear on the paper under `grid_indep_fig/original.`

```python
import numpy
from matplotlib import pyplot, rcParams
%matplotlib notebook
```

```python
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
rcParams['axes.linewidth'] = 1
```

```python
w_19K, cext_19K = numpy.loadtxt('results_data/cube_535nm_res/cube_19K10-20microns.txt',
                        skiprows=1, unpack=True)
w_15K, cext_15K = numpy.loadtxt('results_data/cube_535nm_res/cube_15K10-20microns.txt',
                        skiprows=1, unpack=True)
```

```python
pyplot.figure(figsize=(8, 6))
pyplot.plot((w_19K/5350), cext_19K)
pyplot.xlabel('$\lambda / L$')
pyplot.ylabel('$C_{ext} [nm^2]$')
pyplot.title('19K')
pyplot.grid(True);
```

```python
pyplot.figure(figsize=(8, 6))
pyplot.plot((w_15K/5350), cext_15K)
pyplot.xlabel('$\lambda / L$')
pyplot.ylabel('$C_{ext} [nm^2]$')
pyplot.title('15K')
pyplot.grid(True);
```

## SIC Cube L= 535nm  (Rockstuhl Fig 18 a)

```python
pyplot.figure(figsize=(8, 6))
pyplot.plot((w_19K/5350), cext_19K/(535**2), label='19K', ls='--', lw=2)
pyplot.plot((w_15K/5350), cext_15K /(535**2), label='15K',ls='-',alpha=0.7, lw=2)
pyplot.plot((w_15K/5350), abs(cext_15K /(535**2) - cext_19K /(535**2)), label='diff',ls='-',alpha=0.5, lw=2)

pyplot.xlabel('$\lambda / L$')
pyplot.ylabel('$C_{ext}/L^2 $')
#pyplot.title('15K vs 19K')
#pyplot.xlim(19.25, 19.75)
#pyplot.ylim(0, 500000)
pyplot.legend()
pyplot.grid(True)
pyplot.savefig('grid_indep_fig/cubeL535nm_15Kvs19K.pdf', format='pdf',bbox_inches='tight');
```
